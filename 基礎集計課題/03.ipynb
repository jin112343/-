# PART3 グラフを描いてみる

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## データの読み込み
- train.csvを読み込む
- 読み込んだものは変数trainに代入

```python
train = pd.read_csv("train.csv")
```

## trainの先頭行を確かめてみる

```python
train.head()
```

## trainのyの折れ線グラフを描いてみる
- 折れ線グラフを描くにはplot関数を使う

```python
train["y"].plot()
```

## グラフの大きさをもっと大きくしてみる
- カッコの中にオプションとしてfigsize=(12,4)と書く

```python
train["y"].plot(figsize=(12,4))
```

## グラフにタイトルをつけてみる
- カッコの中にオプションとして、title="XXX"と書く
- カッコの中が複数となった場合はカンマ（,）を使うことで複数書ける

```python
train["y"].plot(figsize=(12,4),title="グラフ")
```

## グラフのx軸とy軸に名前を付けてみる
- ちょっと複雑で一旦、plot関数の結果を別の変数に代入する必要がある
- 今回は変数axに代入することにする
- x軸はaxを使い、set_xlabel("name")と書くことで名前をつける

```python
ax = train["y"].plot(figsize=(12,4),title="グラフ")
ax.set_xlabel("time")
ax.set_ylabel("y")
```

## trainのyのヒストグラムを描いてみる
- グラフを描く前に簡単にヒストグラムとは何かを理解する
- plot.hist関数を使う

```python
train["y"].plot.hist()
```

## グリッド線を引いてみる
- カッコの中にオプションとして、grid=Trueと書く

```python
train["y"].plot.hist(grid=True)
```

## ヒストグラム上に平均値を表す赤線を描く
- ヒストグラムで一番数が多いところが平均値とは限らないケースもある為、ヒストグラムに平均値を表す線を描いてみる
- まず縦線の描き方から始める。縦線を描くときにはplt.axvline(x=NUMBER, color="red")と書く
- vlineのvはvertical（垂直）

```python
plt.axvline(x=2,color="red")
```

```python
plt.axvline(x=train["y"].mean(),color="red")
train["y"].plot.hist()
```

## 上記のヒストグラムを大きさ(12,4)にし、タイトルを"ヒストグラム"として、pngファイルの図として保存
- plt.savefig("FILE_NAME.拡張子")を一番最後に書く
- ファイル名は「sample_fig.png」とする

```python
plt.axvline(x=train["y"].mean(),color="red")
train["y"].plot.hist(figsize=(12,4),title="ヒストグラム")
plt.savefig("sample_fig.png")
```

## yとweekの関係を調べる為に、箱ひげ図を描いてみる
- 箱ひげ図の描き方は、yとweekを選択した上で、boxplot(by="week")と書く
- by="week"はx軸にしたいカラムを選択
- ○は外れ値

```python
train[["y","week"]].boxplot(by="week")
```
